```python
from collections import defaultdict
from abc import ABC, abstractmethod
class Animal(ABC):
    def __init__(self,name,age):
        self.name=name
        self.age=age
    @abstractmethod
    def make_sound(self):
        pass
    def feed(self):
        print(f"{self.name} is being fed")
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def get_type(self):
        return self.__class__.__name__    
class cow(Animal):
    def make_sound(self):
        return "mowwo"
        
    def milk(self):  
        return f"{self.name} giving milk."
        
class chicken(Animal):
    def make_sound(self):
        return "pakpak"  
        
    def lay_egg(self): 
        return f"{self.name} laid an egg."
        
class sheep(Animal):
    def make_sound(self):
        return "mehemehe" 
    def shear(self):
        return f"{self.name} has been sheared."

class FarmStructure:
    def __init__(self, name, structure_type):
        self.name = name
        self.structure_type = structure_type

    def describe(self):
        return f"{self.name} ({self.structure_type})"

class Farm:
    def __init__(self,name):
        self.__name = name
        self.__animals=[]
        self.__structures=[]
        
    def add_animal(self, animal):
        self.__animals.append(animal)
        
    def remove_animal(self, animal_name):
        self.__animals = [a for a in self.__animals if a.get_name() != animal_name]

    def add_structure(self, structure):
        self.__structures.append(structure)

    def remove_structure(self, structure_name):
        self.__structures = [s for s in self.__structures if s.name != structure_name]

    def show_population(self):
        print(f"Welcome to {self.__name}!")
        counter = defaultdict(int)
        for animal in self.__animals:
            counter[animal.get_type()] += 1

        print("Farm Population:")
        for species, count in counter.items():
            print(f"- {species}: {count}")

    def list_structures(self):
        print("Structures:")
        for structure in self.__structures:
            print(structure.describe())
            
    def daily_routine(self):
        print("----- Morning Routine ------")
        products = []
        for animal in self.__animals:
            animal.feed()
            print(f"{animal.get_name()} says: {animal.make_sound()}")

            # Collect products polymorphically
            if isinstance(animal, Cow):
                products.append(animal.milk())
            elif isinstance(animal, Chicken):
                products.append(animal.lay_egg())
            elif isinstance(animal, Sheep):
                products.append(animal.shear())

        print("Collected products:", ", ".join(products))
        
    

```
